# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
```

```python
#Используемый датасет: MovieLens Latest Datasets   ml-latest-small.zip

my_movies = pd.read_table('movies.csv', sep=',', header=0)
my_ratings = pd.read_table('ratings.csv', sep=',', header=0)
movies_merge = pd.merge(my_movies, my_ratings, on='movieId', how='left')
```

```python
#Первый вариант - за счёт мерджа, каждая строка для фильма - это отдельная оценка. Считаем value_counts()
movies_merge[movies_merge['rating'] == 5 ]['title'].value_counts().head(1)
```

```python
#Второй вариант - группировка и список по убыванию
movies_grp = movies_merge[movies_merge['rating'] == 5 ].groupby('title').count()['rating'].sort_values(ascending=False)
movies_grp = pd.DataFrame(movies_grp)
movies_grp.columns = ['rating > 5 (count)']
movies_grp.head()
```

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
#Загрузка данных и проверка названия стран
power = pd.read_csv('power.csv')
power['country'].unique()
```

```python
#Фильтрация данных
cat = [4, 12, 21]
power = power[ power['category'].isin(cat) ] # фильтр по категориям

cnt = ['Lithuania' , 'Latvia', 'Estonia']
power = power[ power['country'].isin(cnt)] #фильтр по странам

power = power[ (power['year']>=2005) & (power['year']<=2010) ] # фильтр по годам

power = power[ power['quantity'] > 0 ] # откидываем отрицательные значения
```

```python
#Первый вариант - суммарное потребление по всем странам одним числом
power['quantity'].sum()
```

```python
#Второй вариант - а-ля сводная таблица с общим итогом
import numpy as np
power_pivot = power[ ['country', 'quantity' ] ]
power_pivot = power_pivot.pivot_table(power_pivot, columns=['country'], aggfunc=np.sum)
power_pivot['total'] = power_pivot.iloc[0,:].sum()
power_pivot
```
